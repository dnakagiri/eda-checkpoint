```python
# I'll change 'Birthday' & 'years played'? to datetime data type from object
gen_stats['Birthday'] = pd.to_datetime(gen_stats['Birthday'])
gen_stats.dtypes
```

```python
# I'll change the following columns from object data type to a numeric data type
# Passes Attempted, Passes Completed, Completion Percentage, Passing Yards, Passing Yards Per Attempt, Passing Yards Per Game,
# TD Passes, Percentage of TDs per Attempts, Ints, Int Rate, Longest Pass, Passes Longer than 20 Yards, 
# Passes Longer than 40 Yards, Sacks, Sacked Yards Lost 

cols = ['Passes Attempted', 'Passes Completed', 'Completion Percentage', 'Passing Yards', 'Passing Yards Per Attempt', 
        'Passing Yards Per Game', 'TD Passes', 'Percentage of TDs per Attempts', 'Ints', 'Int Rate', 'Longest Pass', 
        'Passes Longer than 20 Yards', 'Passes Longer than 40 Yards', 'Sacks', 'Sacked Yards Lost']
passing_stats[cols] = passing_stats[cols].apply(pd.to_numeric, errors='coerce', axis=1)

passing_stats.dtypes
# df = df.astype({"Fee": int, "Discount": float})
```

```python
# I shouldn't have converted all the numerical datatypes above to a float; some of them should be 'int' data types
# so I convert them to int. I use 'Int64' because I wanted to keep the NaN values that couldn't be converted if I used 'int'
passing_stats['Passes Attempted'] = passing_stats['Passes Attempted'].astype('Int64')
passing_stats['Passes Completed'] = passing_stats['Passes Completed'].astype('Int64')
passing_stats['Passing Yards'] = passing_stats['Passing Yards'].astype('Int64')
passing_stats['TD Passes'] = passing_stats['TD Passes'].astype('Int64')
passing_stats['Ints'] = passing_stats['Ints'].astype('Int64')
passing_stats['Longest Pass'] = passing_stats['Longest Pass'].astype('Int64')
passing_stats['Passes Longer than 20 Yards'] = passing_stats['Passes Longer than 20 Yards'].astype('Int64')
passing_stats['Passes Longer than 40 Yards'] = passing_stats['Passes Longer than 40 Yards'].astype('Int64')
passing_stats['Sacks'] = passing_stats['Sacks'].astype('Int64')

passing_stats.dtypes
```

```python
# re-do summary statistics for passing_stats
passing_stats.describe()
```

```python
# creating column that specifies whether position plays offense or defense or special teams
def create_field_position(series):
    if series in ['RB', 'FB', 'LT', 'LG', 'C', 'RG', 'RT', 'WR', 'TE', 'OT', 'G', 'OG', 'T']:
        return "off"
    elif series in ['DE', 'DT', 'LB', 'CB', 'S', 'SS', 'FS', 'DB', 'MLB', 'ILB', 'OLB', 'NT', 'SAF', 'DL']:
        return "dff"
    elif series in ['K', 'LS', 'P']:
        return "special teams"
    elif series == 'QB':
        return "qtrback"
    else:
        return "rtd"
    
gen_stats["field_posn"] = gen_stats["Position"].apply(create_field_position)
gen_stats.head()
```

```python
# creating column that specifies which college football confrence players were in
def create_tourn(series):
    if series in ['Baylor', 'Iowa State', 'Kansas', 'Kansas State', 'Oklahoma', 'Oklahoma State', 'TCU', 'Texas', 'Texas Tech'
              'West Virginia']:
        return "Big12"
    elif series in ['Ohio State', 'Michigan', 'Iowa', 'Penn State', 'Wisconsin', 'Nebraska', 'Minnesota', 
                    'Northwestern', 'Michigan State', 'Illinois', 'Purdue', 'Rutgers', 'Maryland', 'Indiana']:
        return "Big10"
    elif series in ['Alabama', 'Florida', 'Georgia', 'Mississippi', 'Arkansas', 'Texas A&M', 'LSU', 'Tennesse', 'Auburn', 
                    'Missouri', 'South Carolina', 'Kentucky', 'Mississippi State', 'Vanderbilt']:
        return "SEC"
    elif series in ['Washington', 'Washington State', 'Stanford', 'Carlifornia', 'Oregon State', 'Oregon', 'Colorado', 'USC',
                   'Utah', 'Arizona State', 'UCLA', 'Arizona']:
        return "Pac-12"
    elif series in ['Boston College', 'Clemson', 'Florida State', 'Louisville', 'North Carolina State', 'Syracuse',
                   'Wake Forest', 'Duke', 'Georgia Tech', 'Miami', 'North Carolina', 'Pittsburgh', 'Virginia', 'Virginia Tech']:
        return "ACC"
    else:
        return "other"
    
gen_stats["college_conf"] = gen_stats["College"].apply(create_tourn)
gen_stats.head()
```
